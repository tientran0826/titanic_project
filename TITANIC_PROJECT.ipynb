Sử dụng thư viện sklearn hỗ trợ việc gán Label cho cột `Embarked` , và `Sex`

```python
le = LabelEncoder()
le.fit(df_train['Embarked'])
df_train.loc[:,'Embarked'] = le.transform(df_train['Embarked']).copy()
df_test.loc[:,'Embarked']= le.transform(df_test['Embarked']).copy()
```

Kiểm tra dữ liệu ta thấy rằng dữ liệu trong `Embarked` đã chuyển thành số tương ứng, ta giải quyết xong việc chuyển các dữ liệu dạng text sang số để thuận tiện cho việc training Model , tuy nhiên dữ liệu có phạm vị giá trị khá lớn (Cột `Age` và `Fare`) nên ta cần `Chuẩn hoá dữ liệu` về cùng một khoảng giá trị.

Sử dụng phương pháp Standard Scaler (giống việc đưa dữ liệu về phân phối chuẩn)

Công thức: 
- *(x - mean)/std*

Trong đó: 
- x là giá trị từng dòng
- mean là giá trị trung bình cột
- std là độ lệch chuẩn cùa cột

```python
def StdScaler(x,mean,std):
    return (x - mean)/std
```

Tính các giá trị mean và std của từng dữ liệu

```python
age_df_train_std = np.std(df_train.loc[:,'Age'])
age_df_train_mean = np.mean(df_train.loc[:,'Age'])
age_df_test_std = np.std(df_test.loc[:,'Age'])
age_df_test_mean = np.mean(df_test.loc[:,'Age'])
```

```python
df_train.loc[:,'Age'] = df_train.loc[:,'Age'].apply(lambda x: StdScaler(x,age_df_train_mean,age_df_train_std)).copy()
df_test.loc[:,'Age'] = df_test.loc[:,'Age'].apply(lambda x: StdScaler(x,age_df_test_mean,age_df_test_std)).copy()
```

```python
fare_df_train_std = np.std(df_train.loc[:,'Fare'])
fare_df_train_mean = np.mean(df_train.loc[:,'Fare'])

fare_df_test_std = np.std(df_test.loc[:,'Fare'])
fare_df_test_mean = np.mean(df_test.loc[:,'Fare'])
```

```python
df_train.loc[:,'Fare'] = df_train.loc[:,'Fare'].apply(lambda x: StdScaler(x,fare_df_train_mean,fare_df_train_std)).copy()
df_test.loc[:,'Fare'] =df_test.loc[:,'Fare'].apply(lambda x: StdScaler(x,fare_df_test_mean,fare_df_test_std)).copy()
```

### Split dataset

Vì chúng ta chỉ có một dataset có dữ liệu chính xác (được cho trước) nên ta chia dữ liệu thành 2 phần :
- 1 phần để training dữ liệu
- 1 phần cho việc test dữ liệu của model vừa được train

Để thuận tiện sklearn có hỗ trợ module cho việc phân chia dữ liệu

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_train,label,random_state = 40 , test_size = 0.3)
```

- `Random_state` có tác dụng để giữ ổn định dữ liệu khi chạy lại đoạn code
- `Test_size` là số lượng dòng được dùng cho dữ liệu test (ở đây 0.3 là 30%)
- `x_train` là dữ liệu train , `y_train` là đầu ra tương ứng dữ liệu của x_train
- `x_test` là dữ liệu test , `y_test` là đầu ra tương ứng dữ liệu của x_test

Dữ liệu được chia ra ở x_test và x_train sẽ tương tứng 3 phần cho test và 7 phần cho train như đã set ở test_size

## Training model 

Dữ liệu đã được chuẩn bị bao gồm `df_train` và `df_test` , `df_test` là dữ liệu mà bài toán đặt ra model chúng ta cần đoán được , muốn được vậy dữ liệu training của chúng ta phải thực hiện được tốt trên tập `df_train` (dữ liệu thực tế) . Việc chia `df_train` thành tập train và test cũng có tác dụng tương tự , việc chuẩn đoán mô hình vừa sử dụng chạy đúng thực tế hay không đòi hỏi phải có một tập dữ liệu chính xác , tuy nhiên chúng ta chỉ có một tập `df_train` nên cách duy nhất là chia tập trên thành 2 phần.

Việc đầu tiên ta cần thực hiện là cho model chạy gần chính xác trên tập `df_train` sau khi có kết quả gần hài lòng thì ta thực thi trên tập dữ liệu đầu đề `df_test`.

Có thể có nhiều thuật toán xử lý dữ liệu . tuy nhiên mỗi thuật toán cho một độ chính xác đầu ra khác nhau , để kiểm tra dữ liệu dự đoán bằng model và dữ liệu thật , ta sử dụng một module có sẵn trong `sklearn` là `accuracy_score`.

```python
from sklearn.metrics import accuracy_score
```

#### Ta thử với thuật toán đầu tiên là Passive Aggressive Classifier trong thư viện sklearn

```python
from sklearn.linear_model import PassiveAggressiveClassifier
```

Các thao tác sử dụng thư viện ML có thể xem trên youtube Code Explore

```python
pac = PassiveAggressiveClassifier()
pac.fit(x_train,y_train)
```

Ta tiến hành chuẩn đoán dữ liệu đầu ra bằng thuật toán Passive Aggressive Class trên tập dữ liệu x_test

```python
y_pac_pred = pac.predict(x_test)
```

```python
y_pac_pred
```

Ta so sánh dữ liệu vừa dữ đoán và dữ liệu thực tế trong tập `y_test`

```python
pd.DataFrame(y_pac_pred,y_test)
```

Ta thấy có sự khá tương đồng giữa 2 cột dữ liệu , điều ta mong muốn là dữ liệu ta đoán bằng model phải gần chính xác nhất so với dữ liệu thực `y_test` , tuy nhiên nhìn mắt thường rất khó để xác định được việc giống nhau là như thế nào nên ta sử dụng module `accuracy_score` để máy thực hiện so khớp

```python
accuracy_score(y_test,y_pac_pred)
```

Kết quả trả về là 79% (mỗi lần chạy có thể khác nhau vì predict khác nhau) , tuy nhiên ta vẫn muốn tỷ lệ đó cao hơn . Để việc tỷ lệ đó cao hơn thì ta có thể xử lý lại data ban đầu hoặc thay đổi về thuật toán ML đang sử dụng

#### Thuật toán Support Vector Machine trong sklearn

Ta thực hiện tương tự các bước ở trên

```python
from sklearn.svm import SVC

SVM_KERNEL = "linear"
SVM_C = 10
SVM_GAMMA = 0.00001

svm_model = SVC(kernel = SVM_KERNEL, C = SVM_C, gamma = SVM_GAMMA)
svm_model.fit(x_train,y_train)
y_test_pred_2 = svm_model.predict(x_test)
```

```python
accuracy_score(y_test_pred_2,y_test)
```

#### Áp dụng model vào dữ liệu đầu đề 

Trong vài trường hợp SVC trả về kết quả tốt hơn nên ta sử dụng thuật toán này cho dữ liệu đầu đề đưa ra `df_test`

```python
y_pred = svm_model.predict(df_test)
y_pred
```

Ta không được cho kết quả đầu ra nên không thể kiểm tra accuracy , việc xác định accuracy sẽ quyết định số điểm trên Kaggle

Xuất file kết quả theo đúng định dạng file `submission.csv` được cho để làm mẫu

```python
result = pd.DataFrame(df_test.loc[:,'PassengerId'])
```

```python
result['Survived'] = y_pred
```
